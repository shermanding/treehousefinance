```python
def pascal_elements(n):
    # base case
    if n == 0:
        return [1]
    # recursion to return the elements within the arrays of N via for loop
    else:
        N = pascal_elements(n-1)
        return [1] + [N[i] + N[i+1] for i in range(n-1)] + [1]
    
# function to print the elements of pascal's triangle
def pascal(n):
    for i in range(n):
        print(pascal_elements(i))
```

```python
# test function
pascal(5)
```
