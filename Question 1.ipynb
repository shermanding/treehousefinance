```python
import pandas as pd
import numpy as np
```

### 1a

```python
id = [1,2,3,4,5,6,7]
Name = ['John','Mike','Sally','Jane','Joe','Dan','Phil']
Salary = [300,200,550,500,600,600,550]
manager_id = [3,3,4,7,7,3,'']

employee_table = pd.DataFrame({'id':id, 'Name':Name, 'Salary':Salary, 'manager_id':manager_id}).set_index('id')
employee_table
```

```python
# Ignore the last row, because Phil does not have a manager
# Adjust manager_id for 0 index
for i in range(len(Salary)-1):
    if Salary[i] > Salary[manager_id[i]-1]:
        print(Name[i])
```

### 1b

```python
managers = []

for i in range(len(manager_id)-1):
    managers.append(Name[manager_id[i]-1])
set(managers)
```

*We can see that non-managers are John, Mike, Doe and Dan*

```python
non_managers = pd.DataFrame()
non_managers = non_managers.append(employee_table.loc[employee_table['Name'] == 'John'])
non_managers = non_managers.append(employee_table.loc[employee_table['Name'] == 'Mike'])
non_managers = non_managers.append(employee_table.loc[employee_table['Name'] == 'Joe'])
non_managers = non_managers.append(employee_table.loc[employee_table['Name'] == 'Dan'])
non_managers
```
